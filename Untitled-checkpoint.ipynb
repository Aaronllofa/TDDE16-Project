```python
import numpy as np
import pandas as pd
import spacy

#importing TextBlob for pretrained 
from textblob import TextBlob
```

```python
df_tweets=pd.read_csv("covid-19_vaccine_tweets_with_sentiment.csv",encoding="iso-8859-1")
```

```python
df_tweets.info
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
tokenizer=RegexpTokenizer(r'\w+')
ps=PorterStemmer()
en_stop=set(stopwords.words('english'))
```

```python
def getTextCleaned(text):
    text = text.lower()
    tokens=tokenizer.tokenize(text)
    new_tokens=[token for token in tokens if token not in en_stop]
    stemmed_tokens=[ps.stem(tokens) for tokens in new_tokens]
    clean_text=" ".join(stemmed_tokens)
    return clean_text
```

```python
df_tweets['tweet_text']=df_tweets['tweet_text'].apply(getTextCleaned)
df_tweets['tweet_text'].head()
```

```python
def getSubjectivity(tweet_text):
    return TextBlob(tweet_text).sentiment.subjectivity
```

```python
def getPolarity(tweet_text):
    return TextBlob(tweet_text).sentiment.polarity
```

```python
df_tweets['SubjectivityBlob'] = df_tweets['tweet_text'].apply(getSubjectivity)
df_tweets['PolarityBlob'] = df_tweets['tweet_text'].apply(getPolarity)
```

```python
df_tweets.head()
```

```python
def getAnalysisDB(score):
    if score == 1:
        return 'Negative'
    elif score == 2:
        return 'Neutral'
    else:
        return 'Positive'
```

```python
def getAnalysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
```

```python
df_tweets['AnalysisDB'] = df_tweets['label'].apply(getAnalysisDB)
```

```python
df_tweets['AnalysisBlob'] = df_tweets['PolarityBlob'].apply(getAnalysis)
```

```python
#MAKING THE TTF-IDF MATRIX
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

tf=TfidfVectorizer()
text_tf=tf.fit_transform(df_tweets['tweet_text'].astype('U'))
text_tf
```

```python
from sklearn.model_selection import train_test_split

from sklearn.multiclass import OneVsRestClassifier
from sklearn.multiclass import OneVsOneClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, f1_score
```

```python
#Divide data for models
X=text_tf
y=df_tweets['label']

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
#OVR + LR CLASSIFIER
tuned_parameters = [{'estimator__C': [100, 10, 1, 0.1, 0.01, 0.001, 0.0001], 'estimator__penalty': ["l2"]}]

OvR_clf=OneVsRestClassifier(LogisticRegression())

grid_lr_ovr=GridSearchCV(OvR_clf, tuned_parameters, cv=3, scoring='accuracy', error_score='raise')

grid_lr_ovr.fit(X_train, y_train)
```

```python
y_pred = grid_lr_ovr.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(grid_lr_ovr.best_params_)
```

```python
#ONO + LR CLASSIFIER
tuned_parameters = [{'estimator__C': [100, 10, 1, 0.1, 0.01, 0.001, 0.0001], 'estimator__penalty': ["l2"]}]

OvO_clf=OneVsOneClassifier(LogisticRegression())

grid_lr_ovo=GridSearchCV(OvO_clf, tuned_parameters, cv=3, scoring='accuracy', error_score='raise')

grid_lr_ovo.fit(X_train, y_train)
```

```python
y_pred = grid_lr_ovr.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(grid_lr_ovr.best_params_)
```

```python
#F1 COMPARATION OVR VS OVO
grid_lr_ovr_predictions = grid_lr_ovr.predict(X_test)
ovrlr_f1 = f1_score(y_test, grid_lr_ovr_predictions, average='weighted')
grid_lr_ovo_predictions = grid_lr_ovo.predict(X_test)
ovolr_f1 = f1_score(y_test, grid_lr_ovo_predictions, average='weighted')
```

```python
print("F1 OVR LR:", ovrlr_f1)
print("F1 OVO LR:", ovolr_f1)
```
